```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gp
from sklearn import datasets
import seaborn as sns    
```

```python
df = pd.read_csv(r"C:\Users\prani\Downloads\globalterrorismdb_0718dist.csv",encoding='latin1',low_memory=False)
```

```python
np.shape(df)
```

```python
df.head(30)
```

```python
df1=df.replace(np.nan, 0)
df.head(30)
```

```python
col=['scite1','scite2','scite3','propextent','propextent_txt','addnotes','claimmode_txt']
df1=df.drop(columns=col,axis=1)
df1=df.replace(np.nan, 0)
```

```python
df1.isnull().sum()
```

```python
df1.corr()
```

# Number of terrorist attacks Country wise

```python
print(df1['country_txt'].value_counts())
plt.figure(figsize=(35,20))
sns.countplot(x=df1['country_txt'])
plt.xticks(rotation=90)
plt.show()
```

# Year 2014 has recorded most terrorist attacks

```python
print(df1['iyear'].value_counts())
plt.figure(figsize=(35,20))
sns.countplot(x=df1['iyear'])
plt.xticks(rotation=90)
plt.show()
```

# Common Weapon type

```python
print(df1['weaptype1_txt'].value_counts())
plt.figure(figsize=(35,20))
sns.countplot(x=df1['weaptype1_txt'])
plt.xticks(rotation=90)
plt.show()
```

# Common Target type

```python
print(df1['targtype1_txt'].value_counts())
plt.figure(figsize=(35,20))
sns.countplot(x=df1['targtype1_txt'])
plt.xticks(rotation=90)
plt.show()
```

```python
grouped=pd.DataFrame(df1.groupby(['country_txt','region_txt','provstate','city','weaptype1_txt','attacktype1_txt','gname','motive','iyear'])['eventid','imonth','iday','latitude','longitude','summary','success','targtype1_txt','targsubtype1_txt','corp1','target1','individual','weapsubtype1_txt','weapdetail','nkill','nkillus','nkillter','nwound','nwoundus','nwoundte','property',
'dbsource'].sum().reset_index())
```

```python
df_geo = gp.GeoDataFrame(df1, geometry = gp.points_from_xy(df1.longitude, df1.latitude))
df_geo
```
